```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from tabulate import tabulate

# Load the data
df = pd.read_csv('https://bit.ly/3goOAnt', delimiter=',')

# Extract features (X) and target variable (Y)
X = df.values[:, :-1]
Y = df.values[:, -1]

# Display the DataFrame
print(tabulate(df, headers='keys', tablefmt='psql'))

# Fit a linear regression model
fit = LinearRegression().fit(X, Y)

# Get the coefficients
m = fit.coef_.flatten()
c = fit.intercept_.flatten()
print("m = {0}".format(m))
print("C = {0}".format(c))

# Calculate predictions and residuals
Y_pred = fit.predict(X)
residuals = Y - Y_pred

# Plotting the original data points and the regression line
plt.scatter(X, Y, label='Actual Data')
plt.plot(X, Y_pred, label='Regression Line', color='red')

# Plotting residuals
plt.scatter(X, residuals, label='Residuals', color='green')
plt.axhline(y=0, color='black', linestyle='--', label='Zero Residual Line')

# Draw lines connecting points to residuals
for x_val, res_val in zip(X, residuals):
    plt.plot([x_val, x_val], [res_val, 0], color='red', linestyle='-', linewidth=1)
    plt.annotate(f'Residual: {res_val:.2f}', (x_val, res_val), textcoords="offset points", xytext=(0, 10),
                 ha='center', fontsize=8, color='blue')

# Plotting residuals with data points and their coordinates
for i, (x_val, y_val, residual_val) in enumerate(zip(X, Y, residuals)):
    plt.plot([x_val, x_val], [y_val, y_val - residual_val], color='cyan')  # Cyan lines for residuals
    plt.annotate(f'({x_val[0]:.2f}, {y_val:.2f})', (x_val[0], y_val), textcoords="offset points", xytext=(5, 5),
                 ha='left')

plt.xlabel('#Age of Dog')
plt.ylabel('#Veterinary Visits')

plt.legend()
plt.show()
```
