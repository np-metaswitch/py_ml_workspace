```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as  plt 
```

```python
def plot_speech(wordcloud):
    plt.figure(figsize=(8,8))
    plt.imshow(wordcloud)
    plt.axis("off");
    
```

```python
file_path = r'C:\Users\pedam\Documents\src_data\kaggle_datasets\LibriSpeech\famous_speeches_history\Quit_India_by_Gandhi.txt'  
# Replace with the actual path to your file
# Read the content of the file
with open(file_path, 'r', encoding='utf-8') as file:
    speech = file.read()
```

```python
# Tokenize the text into words
words = speech.split()

# Get the total number of words
total_words = len(words)

print(f'Total number of words in the file: {total_words}')
```

```python
# Generate WordCloud
wordcloud = WordCloud(width=500, height=500, random_state=1, background_color='white', stopwords=STOPWORDS, min_font_size=10).generate(speech)

needed_words = wordcloud.words_
# Print the count of words
word_count = len(needed_words)
print(needed_words)
print(f'Total number of unique words in the WordCloud: {word_count}')
```

```python
print(f'Total number of words in the speech Quit India Speech by Gandhi: {total_words}')
print(f'Total number of unique words in the WordCloud: {word_count}')
```

```python
plot_speech(wordcloud)
```
